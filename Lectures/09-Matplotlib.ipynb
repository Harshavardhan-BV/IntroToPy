### Lineplots

We can call the most basic command "plot".

```python
plt.plot()
```

Now let us plot some numbers. This creates a lineplot where the data points are joined together with lines.

On which axis is it getting plotted?

```python
y = np.array([1, 2, 3, 4])
plt.plot(y)
```

Now let us provide data for our x-axis.

```python
x = np.array([1, 2, 3, 4])
y = np.array([2, 2, 3, 6])
plt.plot(x,y)
```

Now let us add labels.

```python
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

But this plot doesn't have the line we plotted previously. It has to be in the same cell.

```python
plt.plot(x,y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

Similarly, we can have multiple plots in the same figure.

```python
y2 = np.array([2, 5, 10, 5])

plt.plot(x,y)
plt.plot(x,y2)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

Suppose we want to know which line is which, we can provide a label for each. However, it is useless without a legend.

```python
y_axex = np.linspace(0,10,5)

plt.plot(x,y,label='Line 1')
plt.plot(x,y2,label='Line 2')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

Let's say we want our y-axis to have only 5 divisions and between 0 to 10.

```python
y_axex = np.linspace(0,10,5)

plt.plot(x,y,label='Line 1')
plt.plot(x,y2,label='Line 2')

plt.yticks(y_axex)

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

We can also set our scale as log values.

```python
y_log = np.logspace(0,1,5)

plt.plot(x,y,label='Line 1')
plt.plot(x,y2,label='Line 2')

plt.yticks(y_log)

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

Alternatively, we can let matplotlib handle the scale as log.

```python
plt.plot(x,y,label='Line 1')
plt.plot(x,y2,label='Line 2')

plt.yscale('log')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

We can also set custom colours with kwargs.

```python
plt.plot(x,y,label='Line 1',c='r')
plt.plot(x,y2,label='Line 2',color='g')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

By default, matplotlib iterates over the tableu colourmap.

```python
for i in range(11):
    plt.plot(x,x+i,label=i)
plt.legend()
```

The thickness of the lines can be changed.

```python
plt.plot(x,y,label='Line 1',lw=5)
plt.plot(x,y2,label='Line 2',linewidth=10)

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

We can change the line styles.

```python
plt.plot(x,y,label='Line 1',ls='--')
plt.plot(x,y2,label='Line 2',linestyle='-.')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

We can add markers for the datapoints.

```python
plt.plot(x,y,label='Line 1',marker='o')
plt.plot(x,y2,label='Line 2',marker='.')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

We can change the figure size as well.

```python
plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.plot(x,y2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('W  I  D  E')
```

And finally let's save the plots.

```python
plt.plot(x,y,label='Line 1')
plt.plot(x,y2,label='Line 2')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')

plt.savefig('./plot.png')
plt.savefig('./plot.svg')
```

### Scatter plots

In certain cases, it doesn't make sense to use lineplots when your data is not continuous.

```python
x=np.random.random(10)
y=np.random.random(10)+2
y2=np.random.random(10)
```

```python
plt.plot(x,y,label='Line')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

It would be better to look at individual points without connecting them. We can set the linestyle to none and add markers instead.

```python
plt.plot(x,y,label='Scatter',linestyle='',marker='o')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

But instead we can use the scatter function to get a scatter plot.

```python
plt.scatter(x,y,marker='v',label='Scatter')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

This also gives us the option to specify different size for each point. This can add another dimension to your data visualization.

```python
plt.scatter(x,y,s=y2*100,label='Scatter')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

And we can even have different colours for the points. But we need a colour bar to be informative.

```python
plt.scatter(x,y,c=y2,label='Scatter')
plt.colorbar()

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

### Bar plots

Bar plots are used for visualizing categorical data. The height for each bar usually represents the counts or % in that category.

```python
x = np.array([1, 2, 3, 4])
y = np.array([2, 2, 3, 6])
y2 = np.array([2, 5, 10, 5])
```

```python
plt.bar(x,y,label='Bar')

plt.legend()
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('TITLE')
```

```python
x_cat = ['Cat1','Cat2','Cat3','Cat4'] 
plt.bar(x_cat,y,label='Bar')

plt.legend()
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('TITLE')
```

There is also a horizontal version of it.

```python
plt.barh(x_cat,y,label='Bar')

plt.legend()
plt.ylabel('Categories')
plt.xlabel('Counts')
plt.title('TITLE')
```

But suppose we want to have two sets of values....

```python
x_cat = ['Cat1','Cat2','Cat3','Cat4'] 
plt.bar(x_cat,y,label='Bar1')
plt.bar(x_cat,y2,label='Bar2')


plt.legend()
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('TITLE')
```

We can stack the bars over each other.

```python
x_cat = ['Cat1','Cat2','Cat3','Cat4'] 
plt.bar(x_cat,y,label='Bar1')
plt.bar(x_cat,y2,bottom=y,label='Bar2')


plt.legend()
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('TITLE')
```

The bars overlap and we cannot see the data underneath. We unfortunately have to use some "hacks" for it.

```python
x = np.arange(0,4)
w = 0.4

x_cat = ['Cat1','Cat2','Cat3','Cat4'] 
plt.bar(x-w/2,y,width=w,label='Bar1')
plt.bar(x+w/2,y2,width=w,label='Bar2')

plt.xticks(x,labels=x_cat)

plt.legend()
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('TITLE')
```

### Histogram

Histograms are used to display the distribution of data.

```python
y_uni = np.random.random(10000)
y_norm = np.random.normal(0,0.1,10000)
```

The data is divided into bins and the height represents the frequency of datapoints in that bin.

```python
plt.hist(y_uni,label='uniform')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Frequency')
plt.title('Uniform Distribution')
```

```python
plt.hist(y_norm,label='normal')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Frequency')
plt.title('Normal Distribution')
```

We can also change the number of bins to make it "smoother".

```python
plt.hist(y_norm, bins=20,label='normal')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Frequency')
plt.title('Normal Distribution')
```

### Box plots

Box plots are used to summarize distributions into 5 ranges: Median, Q1, Q3, Max & Min.

```python
plt.boxplot(y_uni)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Uniform Distribution')
```

Sometimes we get outliers that are plotted outside the whiskers. Let us plot both the distributions together to compare them.

```python
plt.boxplot([y_uni,y_norm])

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks([1,2],['Uniform','Normal'])
plt.title('Comparison of distributions')
```

### Violin plots

Violin plots are similar to box plots but they visualize the distribution of data instead of being limited to 

```python
plt.violinplot([y_uni,y_norm])

plt.xticks([1,2],['Uniform','Normal'])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Comparison of distributions')
```

### 2.5D plots

Sometimes data in 3 dimensions can be visualised on a 2D plot.

```python
mat = np.random.rand(10,10)
```

We can visualize this matrix as a heatmap.

```python
plt.imshow(mat)
plt.colorbar()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

Or use proper axes.

```python
plt.pcolormesh(mat)
plt.colorbar()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

Now let us look at a 2D function.

```python
X,Y=np.meshgrid(np.linspace(0,10,11),np.linspace(0,10,11))
Z = np.exp(-(X-3)**2 - (Y-5)**2)
```

```python
print(X)
```

```python
plt.pcolormesh(X,Y,Z,cmap='coolwarm')
plt.colorbar()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

As seen above we can specify custom colour maps.

```python
plt.colormaps()
```

Or we can visualize it as contours. Contour lines represent points with the same value.

```python
plt.contour(X,Y,Z,cmap='magma')
plt.colorbar()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

We can also visualize vector fields.

```python
U = X + Y
V = X - Y
```

```python
plt.quiver(X,Y,U,V)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

```python
plt.streamplot(X,Y,U,V)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('TITLE')
```

### Sub-plots

We can have multiple plots in the same figure. When we do `plt.subplots` we get an object for figure and axis each.

```python
fig,ax=plt.subplots(2)
print(type(ax))
```

And let's plot the data. The axes is in the form of numpy array and we can index to plot in that 

```python
fig,ax=plt.subplots(2)

ax[0].plot(x,y)
ax[1].plot(x,y2)
```

Since the X-axis is same between the figures, we can share it. Note that we had to use `set_xlabel` for the axis instead of `x_label` of pyplot.

```python
fig,ax=plt.subplots(2,sharex=True)

ax[0].plot(x,y)
ax[0].set_ylabel('Y-axis')

ax[1].plot(x,y2)
ax[1].set_ylabel('Y-axis')
ax[1].set_xlabel('X-axis')
```

We can also have plots where two different axes share the same x-axis. This is useful when comparing plots with different scales.

```python
fig, ax1 = plt.subplots()

ax1.plot(x,x,label='linear',color='tab:blue')
ax1.set_ylabel('Linear',color='tab:blue')
ax1.set_xlabel('X-axis')
ax2 = ax1.twinx()

ax2.plot(x,x**2,label='square',color='tab:orange')
ax2.set_ylabel('Square',color='tab:orange')
```

And, we can have a 2D grid of subplots. We might need the tight layout to avoid overlap of axes.

```python
fig,ax=plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y**2)
ax[1,1].plot(x,y2**2)
plt.tight_layout()
```
